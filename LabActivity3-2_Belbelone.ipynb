```python
import numpy as np
```

```python
months = ['January', 'February', 'March']
product_categories = ['Electronics', 'Clothing']
geographical_regions = ['North', 'South']
```

```python
sales_data = np.random.randint(1000, 10000, size=(3, 2, 2))
```

```python
print("Sample sales data:")
print(sales_data)
```

```python
sales_by_month = np.sum(sales_data, axis=(1, 2))
```

```python
sales_by_category = np.sum(sales_data, axis=(0, 2))
```

```python
sales_by_region = np.sum(sales_data, axis=(0, 1))
```

```python
print("\nSales by month:")
print(sales_by_month)
```

```python
print("\nSales by product category:")
print(sales_by_category)
```

```python
print("\nSales by geographical region:")
print(sales_by_region)
```
