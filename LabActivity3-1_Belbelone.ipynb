```python
import pandas as pd
```

```python
data = {
    'traveller': ['John', 'Alice', 'Bob', 'Eve', 'Charlie'],
    'departure': ['Davao City', 'Davao City', 'Manila', 'Davao City', 'Davao City'],
    'departure_time': ['2023-01-05', '2023-02-10', '2023-03-15', '2023-01-20', '2023-04-25'],
    'arrival': ['Manila', 'Manila', 'Davao City', 'Manila', 'Manila'],
    'arrival_time': ['2023-01-06', '2023-02-11', '2023-03-16', '2023-01-21', '2023-04-26'],
    'flight': ['PAL123', 'PAL456', 'CebuPac789', 'PAL101', 'PAL222'],
    'fare': [5000, 6000, 7000, 5500, 6200]
}
```

```python
df = pd.DataFrame(data)
```

```python
df_dm = df[(df['departure'] == 'Davao City') & (df['arrival'] == 'Manila')]
df_pa = df_dm[df_dm['flight'].str.startswith('PAL')]
df_pa['departure_time'] = pd.to_datetime(df_pa['departure_time'])
df_pa['month'] = df_pa['departure_time'].dt.month
average_fare_per_month = df_pa.groupby('month')['fare'].mean().reset_index()
```

```python
print("Average fare per month for Philippine Airlines flights from Davao City to Manila:")
print(average_fare_per_month)
```
